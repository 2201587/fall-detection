```python
import cv2
from pathlib import Path
from collections import defaultdict

vid_dir = "fall"
vid_path = Path(vid_dir)

# Find all videos recursively
vid_extensions = ['.mp4', '.avi', '.mov', '.MP4', '.AVI', '.MOV']
vid_files = []
for ext in vid_extensions:
    vid_files.extend(vid_path.rglob(f'*{ext}'))

print(f"Found {len(vid_files)} total videos\n")

fps_counts = defaultdict(list)

for file in vid_files:
    cap = cv2.VideoCapture(str(file))
    fps = round(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps if fps > 0 else 0
    cap.release()
    
    # Get relative path
    rel_path = file.relative_to(vid_path)
    
    fps_counts[fps].append({
        'path': str(rel_path),
        'frames': total_frames,
        'duration': duration
    })

# Print FPS distribution
print("=" * 60)
print("FPS DISTRIBUTION:")
print("=" * 60)
for fps, vids in sorted(fps_counts.items()):
    print(f"\n{fps} fps: {len(vids)} videos ({len(vids)/len(vid_files)*100:.1f}%)")
    print(f"  Example: {vids[0]['path']}")
    print(f"           {vids[0]['frames']} frames, {vids[0]['duration']:.1f}s")
```
